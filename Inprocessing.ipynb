```python
#Create binary label dataset that can be used by bias mitigation algorithms
justice_dataset = BinaryLabelDataset(favorable_label=favorable_label,
                                unfavorable_label=unfavorable_label,
                                df=preprocessed_data,
                                label_names=['first_party_winner'],
                                protected_attribute_names=['majority_vote', 'minority_vote'],
                                unprivileged_protected_attributes=unprivileged_groups)
```

```python
display(Markdown("#### Training Data Details"))
print("shape of the training dataset", justice_dataset.features.shape)
print("Training data favorable label (1 means success, 0 means failure)", justice_dataset.favorable_label)
print("Training data unfavorable label", justice_dataset.unfavorable_label)
print("Training data protected attribute", justice_dataset.protected_attribute_names)
print("Training data privileged protected attribute (1: Majority group, 0: Minority group)",
      justice_dataset.privileged_protected_attributes)
print("Training data unprivileged protected attribute (1: Majority group, 0: Minority group)",
      justice_dataset.unprivileged_protected_attributes)
```

```python
justice_dataset_train, justice_dataset_test = justice_dataset.split([0.9], shuffle=True)
```

```python
# Metric for the original dataset
metric_orig_train = BinaryLabelDatasetMetric(justice_dataset_train,
                                             unprivileged_groups=unprivileged_groups,
                                             privileged_groups=privileged_groups)
display(Markdown("#### Original training dataset"))
print("Train set: Difference in mean outcomes between unprivileged and privileged groups = %f" % metric_orig_train.mean_difference())
metric_orig_test = BinaryLabelDatasetMetric(justice_dataset_test,
                                             unprivileged_groups=unprivileged_groups,
                                             privileged_groups=privileged_groups)
print("Test set: Difference in mean outcomes between unprivileged and privileged groups = %f" % metric_orig_test.mean_difference())
```

```python
min_max_scaler = MaxAbsScaler()
justice_dataset_train.features = min_max_scaler.fit_transform(justice_dataset_train.features)
justice_dataset_test.features = min_max_scaler.transform(justice_dataset_test.features)

metric_scaled_train = BinaryLabelDatasetMetric(justice_dataset_train,
                             unprivileged_groups=unprivileged_groups,
                             privileged_groups=privileged_groups)

display(Markdown("#### Scaled dataset - Verify that the scaling does not affect the group label statistics"))
print("Train set: Difference in mean outcomes between unprivileged and privileged groups = %f" % metric_scaled_train.mean_difference())

metric_scaled_test = BinaryLabelDatasetMetric(justice_dataset_test,
                             unprivileged_groups=unprivileged_groups,
                             privileged_groups=privileged_groups)

print("Test set: Difference in mean outcomes between unprivileged and privileged groups = %f" % metric_scaled_test.mean_difference())

```
